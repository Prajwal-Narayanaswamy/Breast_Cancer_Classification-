**Importing the dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# loading the dataset to dataframe
df = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
# adding the 'target' column to the dataframe
df['label'] = breast_cancer_dataset.target
```

Seperating the features and target

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std =  scaler.fit_transform(X_test)

#The fit_transform() method can be used to preprocess data before training a machine learning model.
# This can help the model to learn more quickly and improve its performance
```

```python
print(X_train_std)
```

**Building the Neural Network**

For building this Neural Netwokr, we will be using TensorFlow and Keras libraries.

```python
# importing TensorFlow and Keras
import tensorflow as tf
tf.random.set_seed(3) # random way of generating weights and parameters of the neural network model
from tensorflow import keras
```

```python
from keras.src.layers.serialization import activation
# setting up the layers of Neural Network

model = keras.Sequential([
                            keras.layers.Flatten(input_shape = (30,)), # convers n x m matrix to 1-D array
                            keras.layers.Dense(20, activation= 'relu'), # Relu - Rectified Linear Unit)
                            keras.layers.Dense(2, activation = 'sigmoid')  # Output layer as only 2 neurons because there are 2 classes in our dataset (0/1)
])
```

```python
# compiling the neural network

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', # since we are just using 2 binary classes, we'll be using 'sparse_categorical_crossentropy
              metrics=['accuracy'])
```

```python
# training the Neural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=30)
```

Visualizing the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Accuracy")
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("Model Loss")
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

**Building the predictive system**

```python
input_data = (13.08,15.71,85.63,520,0.1075,0.127,0.04568,0.0311,0.1967,0.06811,0.1852,0.7477,1.383,14.67,0.004097,0.01898,0.01698,0.00649,0.01678,0.002425,14.5,20.49,96.09,630.5,0.1312,0.2776,0.189,0.07283,0.3184,0.08183)

# change the input data to numpy array
input_data_as_np_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_np_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)


prediction_label = [np.argmax(prediction)]
print(prediction_label)

if prediction_label[0] ==0:
  print("The tumor is Malignant")
else:
  print("The tumor is Benign")
```
